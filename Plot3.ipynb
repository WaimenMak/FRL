```python
trajectory = replay_buffer(1e6)
trajectory.load("non_stationary_envs/trajectory")
```

```python
sccd_val = []
fedavg_val = []
fedprox_val = []
moon_val = []
for i in range(len(trajectory.buffer)):
    s = torch.tensor(trajectory.buffer[i][0], dtype=torch.float).unsqueeze(0)
    a = torch.tensor(trajectory.buffer[i][1], dtype=torch.float).unsqueeze(0)
    q1 = sccd_q.Q1_val(s, a)
    q2 = fedavg_q.Q1_val(s, a)
    q3 = moon_q.Q1_val(s, a)
    q4 = fedprox_q.Q1_val(s, a)
    sccd_val.append(q1.detach().item())
    fedavg_val.append(q2.detach().item())
    moon_val.append(q3.detach().item())
    fedprox_val.append(q4.detach().item())
```
